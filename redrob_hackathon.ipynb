```python
for resume in resumes:
    pieces = [p.strip().lower() for p in resume["raw_skills"].split(",")]
    canonical = [SKILL_ALIASES[p] for p in pieces if p in SKILL_ALIASES]
    resume["clean_skills"] = sorted(set(canonical))

print(f"Loaded {len(resumes)} resumes, {len(jds)} JDs, {len(SKILL_ALIASES)} aliases\n")
for resume in resumes:
    print(f"{resume['name']:20s} -> {resume['clean_skills']}")
```

```python
all_skills = set()
for resume in resumes:
    all_skills.update(resume["clean_skills"])
vocab = sorted(all_skills)

df = {skill: sum(1 for r in resumes if skill in r["clean_skills"]) for skill in vocab}
idf = {skill: math.log(10 / df[skill]) for skill in vocab}

resume_vectors = []
for resume in resumes:
    N = len(resume["clean_skills"])
    skills_set = set(resume["clean_skills"])
    vec = [(1/N) * idf[skill] if skill in skills_set else 0.0 for skill in vocab]
    resume_vectors.append(vec)
```

```python
print(f"Vocabulary size: {len(vocab)}")
print(f"Vocab: {vocab}\n")
print("Arjun's non-zero TF-IDF entries:")
for i, skill in enumerate(vocab):
    if resume_vectors[0][i] > 0:
        print(f"  {skill}: {resume_vectors[0][i]:.4f}")
```

```python
jd_vectors = []
for jd in jds:
    pieces = [s.strip().lower() for s in jd["skills"]]
    canonical = set(SKILL_ALIASES[p] for p in pieces if p in SKILL_ALIASES)
    vec = [1 if skill in canonical else 0 for skill in vocab]
    jd_vectors.append(vec)
```

```python
def cosine(a, b):
    dot = sum(x*y for x, y in zip(a, b))
    na = math.sqrt(sum(x*x for x in a))
    nb = math.sqrt(sum(y*y for y in b))
    return dot / (na * nb) if na > 0 and nb > 0 else 0.0
```

```python
for j, jd in enumerate(jds):
    scores = []
    for i, resume in enumerate(resumes):
        s = round(cosine(resume_vectors[i], jd_vectors[j]), 2)
        scores.append((resume["name"], s))
    scores.sort(key=lambda x: (-x[1], x[0]))
    top3 = scores[:3]
    print(f"JD-{j+1} — {jd['company']} ({jd['role']})")
    print(", ".join(f"{name}({score:.2f})" for name, score in top3))
    print()
```
